## 1. Understanding the Basics

**Probability Definition:**
Probability measures the chance that an event will occur.

Formula:
P(Event) = Number of Favorable Outcomes / Total Outcomes

**Examples from Dataset:**
1. Event A: Student passes the exam.
2. Event B: Attendance > 80%.
3. Event C: Participation in group discussion.

```python
import pandas as pd
df = pd.read_csv('expectation_decider_dataset_v2.csv')
df.head()
```

### Empirical Probability of Passing
Total Students = 200
Number of Pass = 64
P(Pass) = 64/200 = 0.32

## 2. Types of Probability

**Empirical Probability:** Calculated from actual dataset.
P(Pass) = 0.32

**Theoretical Probability (Binomial Model):**
Used when selecting 3 students randomly.
X = Number of students passing out of 3.
Follows Binomial Distribution:
P(X = k) = nCk * p^k * (1-p)^(n-k)

### Probability Distribution Table (n = 3)
P(X=0) = 0.3144
P(X=1) = 0.4439
P(X=2) = 0.2089
P(X=3) = 0.0328

Mean = np = 3 × 0.32 = 0.96
Variance = np(1-p) = 0.6528

## 3. Contingency Table
Relationship between Group Discussion and Exam Result.

```python
pd.crosstab(df['group_discussion'], df['final_exam_pass'])
```

## 4. Conditional Probability

We calculate probability of passing given attendance > 80%.

P(High Attendance) = 0.335
P(Pass | High Attendance) = 0.6119

## 5. Bayes Theorem Application

Bayes Formula:
P(Pass | High Attendance) = [P(High Attendance | Pass) × P(Pass)] / P(High Attendance)

Values substituted from dataset are calculated above.

## 6. Venn Diagram (Conceptual Representation)
Study Hours > 10 and Attendance > 80% overlap.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

high_study = len(df[df['study_hours'] > 10])
high_att = len(df[df['attendance'] > 80])
both = len(df[(df['study_hours'] > 10) & (df['attendance'] > 80)])

plt.figure()
c1 = Circle((0,0),2)
c2 = Circle((1.5,0),2)
plt.gca().add_patch(c1)
plt.gca().add_patch(c2)
plt.text(-1.5,0,f'Study>10\n{high_study}')
plt.text(2.5,0,f'Attendance>80%\n{high_att}')
plt.text(0.75,0,f'Both\n{both}')
plt.xlim(-4,5)
plt.ylim(-3,3)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Venn Diagram Representation')
plt.show()
```

- Participation in group discussion shows positive association with results.
